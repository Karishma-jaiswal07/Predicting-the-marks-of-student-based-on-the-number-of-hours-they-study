```python
#NOW WE HAVE PREPARED DATA FOR OUR MODEL.
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
# print(X)
```

```python
#NOW WE ARE DIVIDING OUR DATA-SET FOR TRAINING AND TESTING.
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0)
```

```python
#NOW WE ARE TRAINING OUR DATA USING LINEAR REGRESSION ALGORITHM.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 


# from sklearn.ensemble import RandomForestRegressor
# regressor = RandomForestRegressor(n_estimators = 1000, random_state = 42)


regressor.fit(X_train, y_train) 
print("Training complete.")
```

```python
#NOW LETS TEST OUR DATAS.
print(X_test) 
print("Predection of Score")
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
#NOW LETS CHECK THE ACCURACY OF THE MODEL.
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  

df 
```

```python
#NOW LETS CHECK WITH THE CUSTOM INPUT.
hours = [[9.25]]
pred = regressor.predict(hours)
print(pred)
```

```python
# EVALUATING THE MODEL.
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
